# Quantidade de pessoas por faixa etaria

```python
lista = ['NU_INSCRICAO','TP_FAIXA_ETARIA']
```

```python
novo_df = df.filter(items = lista)
faixa_etaria = novo_df.groupby("TP_FAIXA_ETARIA").count()
faixa_etaria_att = faixa_etaria.reset_index().rename(columns={'NU_INSCRICAO':'quantidade'})
```

# Quantidade de pessoas por sexo

```python
# montadno a selecao 
lista = ['NU_INSCRICAO','TP_SEXO']
# colhendo as informacoes
novo_df = df.filter(items = lista)
quant_sexo = novo_df.groupby("TP_SEXO").count()
quant_sexo = quant_sexo.reset_index().rename(columns={'NU_INSCRICAO':'Quantidade'})
# tirando a porcentagem
fem = quant_sexo.query("TP_SEXO == 'F'")
masc = quant_sexo.query("TP_SEXO == 'M'")
porcentagem_fem = (int(fem['Quantidade'])/len(df['TP_SEXO'])) * 100
porcentagem_masc = (int(masc['Quantidade'])/len(df['TP_SEXO'])) * 100
# adiconando a coluna de porcentagem
col_porencetagem = [round(porcentagem_fem, 2), round(porcentagem_masc, 2)]
quant_sexo['Porcentagem'] = col_porencetagem
quant_sexo
```

# Quantidade por cor/raça

```python
filtro = ['TP_COR_RACA', 'NU_INSCRICAO']
db = df.filter(items = filtro)
db_groupby = db.groupby('TP_COR_RACA').count()
db_groupby = db_groupby.rename(columns = {'NU_INSCRICAO':'Quantidade'}).reset_index()
# adicionando coluna
col = {0:'Nao Declarado', 1:'Branca', 2:'Preta', 3:'Parda', 4:'Amarela', 5:'Indigena',6:'Sem Info'}
db_groupby['Cor'] = db_groupby['TP_COR_RACA'].map(col)
db_groupby
```
